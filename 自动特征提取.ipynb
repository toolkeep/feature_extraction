```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import os

#
def readData(filePath):
    # 
    columnNames = ['user_id', 'activity', 'timestamp', 'x-axis', 'y-axis', 'z-axis']
    data = pd.read_csv(filePath, header=None, names=columnNames, na_values=';')
    return data

# 
def features(data_input):
    acc_avg = np.mean(data_input, axis=0)
    acc_max = np.max(data_input, axis=0)
    acc_min = np.min(data_input, axis=0)
    acc_std = np.std(data_input, axis=0)
    acc_var = np.var(data_input, axis=0)
    acc_zero= len(data_input[data_input>0])
    acc_range = acc_max - acc_min
    return acc_avg, acc_max, acc_min, acc_std, acc_var, acc_zero, acc_range

# 
def windows(data, size):
    start = 0
    while start < data.count():
        yield int(start), int(start + size)     #size
        start += (size/2)    #50%

x_path = r'D:\Item\Laboratory\PYTHON/test_f.csv'  #
dataset = readData(r'D:\Item\Laboratory\PYTHON/actitracker_raw.csv')   #

feature = np.empty((1,22))   #xyz*721+1
all_data = np.empty((1,22,0))   #xyz*721+1
labels= np.empty((1))        #activity

object = windows(dataset['activity'],90)   #size9090*3270

#
for i in object:

    acc_x = dataset['x-axis'][i[0]:i[1]]     #
    acc_y = dataset['y-axis'][i[0]:i[1]]
    acc_z = dataset['z-axis'][i[0]:i[1]]
    
    labels = stats.mode(dataset['activity'][i[0]:i[1]])[0][0]    #activity
    arr = np.reshape(np.array(labels.split(',')),(1,1))        #

    x_feature = pd.DataFrame(features(acc_x)).T   #DataFrame
    y_feature = pd.DataFrame(features(acc_y)).T
    z_feature = pd.DataFrame(features(acc_z)).T
    
    feature = np.hstack([arr, x_feature, y_feature, z_feature])  #
    all_data = np.dstack([all_data,feature])    #
    output_data = np.reshape(all_data,(22,all_data.shape[2]))   #
    output_data = pd.DataFrame(output_data).T     #
    
    print(output_data.shape)  #

#
#
output_data.columns=["activity","x_avg","x_max","x_min","x_std","x_var","x_zero","x_range","y_avg","y_max","y_min","y_std","y_var","y_zero","y_range","z_avg","z_max","z_min","z_std","z_var","z_zero","z_range"]
output_data.index = range(1,len(output_data)+1)    #ID1
output_data.to_csv(x_path,sep=',',encoding="gbk")   #
```

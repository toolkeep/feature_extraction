```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import os

#读取你要提取特征的时序信号文件
def readData(filePath):
    # 数据集的属性
    columnNames = ['user_id', 'activity', 'timestamp', 'x-axis', 'y-axis', 'z-axis']
    data = pd.read_csv(filePath, header=None, names=columnNames, na_values=';')
    return data

# 定义一个时域特征计算函数，需要其他特征比如频域特征可自己添加
def features(data_input):
    acc_avg = np.mean(data_input, axis=0)
    acc_max = np.max(data_input, axis=0)
    acc_min = np.min(data_input, axis=0)
    acc_std = np.std(data_input, axis=0)
    acc_var = np.var(data_input, axis=0)
    acc_zero= len(data_input[data_input>0])
    acc_range = acc_max - acc_min
    return acc_avg, acc_max, acc_min, acc_std, acc_var, acc_zero, acc_range

# 定义一个窗口函数
def windows(data, size):
    start = 0
    while start < data.count():
        yield int(start), int(start + size)     #生成器函数，返回每一段窗口数据的起始和结尾序号，size为窗口的大小
        start += (size/2)    #每个窗口数据与相邻窗口的重叠率，这里为50%的重叠率

x_path = r'D:\Item\Laboratory\PYTHON/test_f.csv'  #保存地址
dataset = readData(r'D:\Item\Laboratory\PYTHON/actitracker_raw.csv')   #读取文件地址

feature = np.empty((1,22))   #xyz三个轴的加速度*7，共有21个特征+1列标签
all_data = np.empty((1,22,0))   #xyz三个轴的加速度*7，共有21个特征+1列标签
labels= np.empty((1))        #标签，对应activity

object = windows(dataset['activity'],90)   #size的大小为90，即每个窗口有90*3共270个时序数据

#遍历生成器对象，对每一个生成器对象对应的数据提取特征，每循环一次叠加一次，最后将所有特征保存至本地
for i in object:

    acc_x = dataset['x-axis'][i[0]:i[1]]     #提取一个窗口大小的时序数据
    acc_y = dataset['y-axis'][i[0]:i[1]]
    acc_z = dataset['z-axis'][i[0]:i[1]]
    
    labels = stats.mode(dataset['activity'][i[0]:i[1]])[0][0]    #提取一个窗口数据中出现最多次的activity作为该次特征集的标签
    arr = np.reshape(np.array(labels.split(',')),(1,1))        #转换为数组格式

    x_feature = pd.DataFrame(features(acc_x)).T   #调用特征计算函数并转换为DataFrame格式
    y_feature = pd.DataFrame(features(acc_y)).T
    z_feature = pd.DataFrame(features(acc_z)).T
    
    feature = np.hstack([arr, x_feature, y_feature, z_feature])  #标签和特征水平方向叠加
    all_data = np.dstack([all_data,feature])    #沿深度方向进行拼接特征
    output_data = np.reshape(all_data,(22,all_data.shape[2]))   #转换为二维输出
    output_data = pd.DataFrame(output_data).T     #转换格式
    
    print(output_data.shape)  #查看目前特征的叠加状态

#退出循环后再保存所有特征
#定义每一列的特征名称（第一列为标签名称）
output_data.columns=["activity","x_avg","x_max","x_min","x_std","x_var","x_zero","x_range","y_avg","y_max","y_min","y_std","y_var","y_zero","y_range","z_avg","z_max","z_min","z_std","z_var","z_zero","z_range"]
output_data.index = range(1,len(output_data)+1)    #ID从1开始递增
output_data.to_csv(x_path,sep=',',encoding="gbk")   #保存到本地
```
